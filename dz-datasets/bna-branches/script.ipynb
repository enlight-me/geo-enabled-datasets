```
import requests 
import json 
import time
import pandas as pd
```

```
full_url = "https://www.bna.dz/fr/index.php?option=com_kdagences&task=getAgences&tmpl=component&lang=fr-FR&wilaya=11"
url = 'https://www.bna.dz/fr/index.php'
params = {'option' : 'com_kdagences' ,'task':'getAgences' ,'tmpl':'component' ,'lang':'fr-FR','wilaya':'1'}

headers = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36"}

wilaya = 1
result = []
for wilaya in range(1,49):
  params['wilaya'] = str(wilaya)
  r = requests.get(url,headers=headers,verify=False, params=params)
  response = r.json()
  
  wilaya_name= response['wilaya']['nom']
  agencies = response['agences']

  for agency in agencies:
    agency.pop('wilaya')
    agency['code_wilaya'] = str(wilaya)
    agency['wilaya'] = wilaya_name
    result.append(agency)
  time.sleep(1)




```

```
df = pd.DataFrame.from_dict(result)
df = df.astype(str)
df = df.replace(r',\r\n$','', regex=True) 
df =df.replace(r',\r\n',' - ',regex =True)
df = df.drop(columns=['id','email','zoom'])
df.rename(columns={'descr':'email'},inplace=True)
df
```
