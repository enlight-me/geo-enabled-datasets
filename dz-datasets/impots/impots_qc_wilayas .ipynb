```python
import pandas as pd 
import geopandas as gpd
from shapely.geometry import Point
import numpy as np
import os
```

```python
home_path=os.getenv('HOME')
wilayas_shp= home_path+'/GeoData/Others/LimAdmin/wilayas.shp'
wilayas_gdf = gpd.read_file(wilayas_shp)
wilayas_gdf.rename(columns={"CODE": "code_wil"}, inplace=True)
wilayas_gdf['centroid'] = wilayas_gdf['geometry'].centroid
wilayas_gdf
```

```python
impots_branches = pd.read_csv('./impots.csv')
impots_branches.head()
```

```python
impots_wilaya = impots_branches[['wilaya']].drop_duplicates()
impots_wilaya.sort_values('wilaya',inplace=True)
```

```python
import fuzzy_pandas as fpd


merge= fpd.fuzzy_merge(impots_wilaya, wilayas_gdf,
                left_on=['wilaya'],
                right_on=['WILAYA'],
                method='jaro',
                ignore_case=True,
                threshold=0.9,
                keep_right=['WILAYA','code_wil'])
merge.drop(25,inplace=True)
```

```python
impots_branches = pd.merge(impots_branches,merge[['wilaya','code_wil']] , on='wilaya')
```

```python
impots_branches.rename(columns={"code_wil":"code"},inplace=True)
```

```python
impots_branches['geometry'] = impots_branches.apply(lambda row:Point(row.lon,row.lat) ,axis =1)
impots_gdf = gpd.GeoDataFrame(impots_branches)
impots_gdf.head()
```

```python
# Set the Coordinate reference system
impots_gdf = impots_gdf.set_crs("EPSG:4326")
wilayas_gdf = wilayas_gdf.set_crs("EPSG:4326")
```

```python
# Update the Targeted DataFrame by adding Wilayas geometries and centroids 
impots_gdf = impots_gdf.merge(wilayas_gdf,left_on='code',right_on='code_wil')
impots_gdf.rename(columns={"geometry_x": "geometry"}, inplace=True)
```

```python
impots_gdf = impots_gdf.sort_values(by=['code'],ignore_index=True)
impots_gdf.reset_index(drop=True)
```

```python
impots_gdf['coords_in_wil']=[impots_gdf.geometry[i].within(impots_gdf.geometry_y[i]) for i in range(impots_gdf.geometry.count())]
impots_gdf[impots_gdf.coords_in_wil==False]
```

```python
final_gdf = gpd.GeoDataFrame(impots_gdf)
final_gdf['to_check'] = (impots_gdf['to_check']==True) | (impots_gdf['coords_in_wil'] ==False)
```

```python
final_gdf
```

```python
# Write to a GeoPackage file
final_gdf = final_gdf.set_crs("EPSG:4326")
final_gdf = final_gdf.drop(['geometry_y', 'code_wil', 'centroid', 'WILAYA', 'coords_in_wil'], axis=1)

gpd.GeoDataFrame(final_gdf).to_file("impots_to_check.geojson", driver='GeoJSON')
```
